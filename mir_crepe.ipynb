```python
import h5py
import numpy as np
import pretty_midi
import os

# Path to your HDF5 file
h5_file_path = "/Volumes/T7/mini_MedleyDB_pitch/processed_pitch_h5/processed_pitches.h5"
output_dir = "/Volumes/T7/mini_MedleyDB_pitch/piano_rolls_crepe"
os.makedirs(output_dir, exist_ok=True)

# Parameters for piano roll
fs = 100  # Sampling frequency for time resolution (100 frames per second)
midi_min_pitch = 21  # Lowest MIDI note (A0)
midi_max_pitch = 108  # Highest MIDI note (C8)

def pitch_to_piano_roll(time, frequency, confidence, fs, confidence_threshold=0.5, window_size=5):
    # Filter out low-confidence predictions
    valid_indices = confidence > confidence_threshold
    time = time[valid_indices]
    frequency = frequency[valid_indices]
    
    # Handle the case where the filtered data is empty
    if len(time) == 0:
        print("No valid data after filtering by confidence threshold.")
        return np.zeros((midi_max_pitch - midi_min_pitch + 1, 1))  # Return an empty piano roll

    # Smooth the frequency values using a moving average filter
    smoothed_frequency = np.convolve(frequency, np.ones(window_size) / window_size, mode='same')

    # Convert smoothed frequency to MIDI note numbers
    midi_notes = np.round(69 + 12 * np.log2(smoothed_frequency / 440.0)).astype(int)
    
    # Clip notes to the valid MIDI range
    midi_notes = np.clip(midi_notes, midi_min_pitch, midi_max_pitch)
    
    # Create time bins
    duration = time.max()
    n_bins = int(np.ceil(duration * fs)) + 1  # Add 1 to ensure edge cases fit
    piano_roll = np.zeros((midi_max_pitch - midi_min_pitch + 1, n_bins))
    
    # Map MIDI notes to piano roll
    for t, note in zip(time, midi_notes):
        time_idx = int(np.floor(t * fs))
        note_idx = note - midi_min_pitch
        if time_idx < n_bins:  # Ensure the index is within bounds
            piano_roll[note_idx, time_idx] = 1  # Set note active
    
    return piano_roll


# Function to save piano roll as MIDI
def piano_roll_to_midi(piano_roll, fs, output_path):
    midi = pretty_midi.PrettyMIDI()
    piano = pretty_midi.Instrument(program=0)  # Acoustic Grand Piano
    
    for pitch, row in enumerate(piano_roll):
        note_on = None
        for time_idx, value in enumerate(row):
            if value > 0 and note_on is None:
                note_on = time_idx
            elif value == 0 and note_on is not None:
                start_time = note_on / fs
                end_time = time_idx / fs
                midi_note = pretty_midi.Note(
                    velocity=100,
                    pitch=pitch + midi_min_pitch,
                    start=start_time,
                    end=end_time
                )
                piano.notes.append(midi_note)
                note_on = None
        if note_on is not None:  # Handle note at the end
            start_time = note_on / fs
            end_time = len(row) / fs
            midi_note = pretty_midi.Note(
                velocity=100,
                pitch=pitch + midi_min_pitch,
                start=start_time,
                end=end_time
            )
            piano.notes.append(midi_note)
    
    midi.instruments.append(piano)
    midi.write(output_path)


# Main loop: Process HDF5 data and convert to piano roll
with h5py.File(h5_file_path, "r") as h5_file:
    for group_name in h5_file.keys():
        print(f"Processing group: {group_name}")
        time = h5_file[group_name]["time"][:]
        frequency = h5_file[group_name]["frequency"][:]
        confidence = h5_file[group_name]["confidence"][:]
        
        # Convert to piano roll
        piano_roll = pitch_to_piano_roll(time, frequency, confidence, fs)
        
        # Save piano roll as HDF5
        piano_roll_path = os.path.join(output_dir, f"{group_name}_piano_roll.h5")
        with h5py.File(piano_roll_path, "w") as piano_h5:
            piano_h5.create_dataset("piano_roll", data=piano_roll)
            print(f"Saved piano roll to: {piano_roll_path}")
        
        # Save as MIDI
        midi_path = os.path.join(output_dir, f"{group_name}.mid")
        piano_roll_to_midi(piano_roll, fs, midi_path)
        print(f"Saved MIDI file to: {midi_path}")
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

def visualize_piano_roll(piano_roll_input, fs, midi_min_pitch=21, midi_max_pitch=108):
    """
    Visualize a piano roll matrix or load it from an HDF5 file.

    Parameters:
    - piano_roll_input: Either a 2D numpy array (piano roll) or a file path (HDF5).
    - fs: Sampling frequency (time bins per second).
    - midi_min_pitch: Lowest MIDI pitch in the piano roll.
    - midi_max_pitch: Highest MIDI pitch in the piano roll.
    """
    # Check if the input is a file path
    if isinstance(piano_roll_input, str):  # If it's a string, assume it's a file path
        with h5py.File(piano_roll_input, "r") as h5_file:
            piano_roll = h5_file["piano_roll"][:]  # Load the piano roll dataset
    else:
        piano_roll = piano_roll_input  # Otherwise, assume it's already a numpy array

    # Generate time and pitch labels
    num_time_bins = piano_roll.shape[1]
    time = np.arange(num_time_bins) / fs
    midi_pitches = np.arange(midi_min_pitch, midi_max_pitch + 1)
    
    # Plot piano roll
    plt.figure(figsize=(12, 6))
    plt.imshow(
        piano_roll, 
        aspect='auto', 
        origin='lower', 
        cmap='viridis', 
        extent=[time[0], time[-1], midi_pitches[0], midi_pitches[-1]]
    )
    plt.colorbar(label="Activation")
    plt.xlabel("Time (s)")
    plt.ylabel("MIDI Pitch")
    plt.title("Piano Roll")
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()
```

```python
import os
import librosa
import h5py
import numpy as np
from tqdm import tqdm

# Load directories
audio_dir = "/Volumes/T7/mini_MedleyDB_pitch/audio"
output_dir = "/Volumes/T7/mini_MedleyDB_pitch/processed_pitch_Pyin_h5"
os.makedirs(output_dir, exist_ok=True)

# Path for the HDF5 file
h5_file_path = os.path.join(output_dir, "processed_pitches_pyin.h5")

# Helper function to extract pitch using PYin
def extract_pitch_pyin(audio, sr):
    """
    Extract pitch using Librosa's PYin algorithm.
    """
    pitches, voiced_flags, voiced_probs = librosa.pyin(
        audio,
        fmin=librosa.note_to_hz('C2'),  # Minimum pitch (C2 ~ 65.41 Hz)
        fmax=librosa.note_to_hz('C7'),  # Maximum pitch (C7 ~ 2093 Hz)
        sr=sr,
        fill_na=0.0  # Replace unvoiced frames with 0.0
    )
    # Generate time array for the pitch values
    times = librosa.times_like(pitches, sr=sr)
    return times, pitches, voiced_probs

# Get list of valid audio files (filter out hidden files)
audio_files = [
    os.path.join(audio_dir, f)
    for f in os.listdir(audio_dir)
    if f.endswith('.wav') and not f.startswith('._')
]

# Create and process the HDF5 file
with h5py.File(h5_file_path, "w") as h5_file:
    for audio_path in tqdm(audio_files, desc="Processing audio files"):
        try:
            print(f"Processing: {audio_path}")
            
            # Load audio
            audio, sr = librosa.load(audio_path, sr=16000, mono=True)

            # Run pitch detection using PYin
            time, frequency, confidence = extract_pitch_pyin(audio, sr)

            # Handle cases where PYin returns NaN for unvoiced sections
            frequency = np.nan_to_num(frequency, nan=0.0)
            confidence = np.nan_to_num(confidence, nan=0.0)

            # Store results in HDF5 file
            group_name = os.path.basename(audio_path).replace('.wav', '')
            group = h5_file.create_group(group_name)
            group.create_dataset("time", data=time)
            group.create_dataset("frequency", data=frequency)
            group.create_dataset("confidence", data=confidence)

            print(f"Saved pitch data to HDF5 group: {group_name}")

        except Exception as e:
            # Log errors to a file
            error_log_path = os.path.join(output_dir, "error_log.txt")
            with open(error_log_path, "a") as error_log:
                error_log.write(f"Error processing {audio_path}: {e}\n")
            print(f"Error processing {audio_path}: {e}")
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

def visualize_pyin_piano_roll(piano_roll_input, fs, midi_min_pitch=21, midi_max_pitch=108):
    """
    Debug and visualize a piano roll matrix generated from PYin or load it from an HDF5 file.

    Parameters:
    - piano_roll_input: Either a 2D numpy array (piano roll) or a file path (HDF5).
    - fs: Sampling frequency (time bins per second).
    - midi_min_pitch: Lowest MIDI pitch in the piano roll.
    - midi_max_pitch: Highest MIDI pitch in the piano roll.
    """
    # Load the piano roll
    if isinstance(piano_roll_input, str):  # If it's a string, assume it's a file path
        with h5py.File(piano_roll_input, "r") as h5_file:
            piano_roll = h5_file["piano_roll"][:]  # Load the piano roll dataset
    else:
        piano_roll = piano_roll_input  # Otherwise, assume it's already a numpy array

    # Debugging: Print stats
    print("Piano Roll Shape:", piano_roll.shape)
    print("Piano Roll Data (min, max):", piano_roll.min(), piano_roll.max())

    if piano_roll.max() == 0:
        print("Piano roll is empty or contains only zeros.")
        return

    # Normalize data for better visibility
    piano_roll_normalized = piano_roll / piano_roll.max()

    # Generate time and pitch labels
    num_time_bins = piano_roll.shape[1]
    time = np.arange(num_time_bins) / fs
    midi_pitches = np.arange(midi_min_pitch, midi_max_pitch + 1)
    
    # Plot piano roll
    plt.figure(figsize=(12, 6))
    plt.imshow(
        piano_roll_normalized, 
        aspect='auto', 
        origin='lower', 
        cmap='viridis',  # Better colormap for low values
        extent=[time[0], time[-1], midi_pitches[0], midi_pitches[-1]]
    )
    plt.colorbar(label="Normalized Activation")
    plt.xlabel("Time (s)")
    plt.ylabel("MIDI Pitch")
    plt.title("Piano Roll (PYin)")
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()
```
